```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.columns.values
```

```python
data = data.drop(['PassengerId', 'Name' , 'Ticket' , 'Cabin'],axis=1)
data
```

```python
data['Age'].fillna(data.Age.median() , inplace =True)
```

```python
data.Embarked.value_counts()
```

```python
mean_fare= data.Fare.mean()
data.Fare.fillna(mean_fare, inplace = True)
```

```python
mode_Embarked = data.Embarked.mode()[0]
data.Embarked.fillna(mode_Embarked, inplace = True)
```

```python
median_age = data.Age.median()
data.Age.fillna(median_age, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.Sex.replace({
'male':1,
 'female':0
    } ,inplace =True)
```

```python
data.tail()
```

```python
data['Embarked_Q'] = (data.Embarked=='Q').astype(int)
data['Embarked_C'] = (data.Embarked=='C').astype(int)
data['Embarked_S'] = (data.Embarked=='S').astype(int)
```

```python
data = data.drop('Embarked', axis=1)
```

```python
data.head()
```

# Standaradization

```python
mean_age = data['Age'].mean()
std_age = data['Age'].std()
data.Age = (data['Age']-mean_age)/std_age
```

```python
mean_fare = data['Fare'].mean()
std_fare = data['Fare'].std()
data.Fare = (data['Fare']-mean_fare)/std_fare
```

```python
mean_parch = data['Parch'].mean()
std_parch = data['Parch'].std()
data.Parch = (data['Parch']-mean_parch)/std_parch
```

```python
mean_SibSp = data['SibSp'].mean()
std_SibSp = data['SibSp'].std()
data.SibSp = (data['SibSp']-mean_SibSp)/std_SibSp
```

```python
data.head()
```

# Shuffle the data

```python
scaled_data = data.copy()
```

```python
suffled_indices =  np.arange(scaled_data.shape[0])
suffled_indices
```

```python
np.random.seed(333334)
np.random.shuffle(suffled_indices)
suffled_indices
```

```python
suffled_scaled_data = scaled_data.iloc[suffled_indices]
```

```python
suffled_scaled_data.head()
```

# Create Test set

```python
test_ratio = 0.15
num_all_rows , num_cols = suffled_scaled_data.shape
```

```python
num_test_rows = int(test_ratio *num_all_rows)
num_train_rows = int(num_all_rows - num_test_rows)
```

```python
train_set = suffled_scaled_data[:num_train_rows]
test_set = suffled_scaled_data[num_train_rows:]
```

```python
train_set.to_csv('final_train_set.csv')
```

```python
test_set.to_csv('final_test_set.csv')
```
